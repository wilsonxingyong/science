```python
import pyecharts.options as opts
from pyecharts.charts import Bar, Line

"""
Gallery 使用 pyecharts 1.0.0
参考地址: https://www.echartsjs.com/examples/editor.html?c=multiple-y-axis

目前无法实现的功能:

1、暂无
"""

colors = ["#5793f3", "#d14a61", "#675bba"]
x_data = ["海藻酸鈉溶液0.5%", "海藻酸鈉溶液0.8%", "海藻酸鈉溶液1%", "海藻酸鈉溶液1.2%", "海藻酸鈉溶液1.5%"]
legend_list = ["第一次伸長量", "第二次伸長量", "第三次伸長量", "第四次伸長量", "第五次伸長量"]
elongation1_capacity = [
    1.1,
    0.6,
    1.6,
]
elongation2_capacity = [
    1.0,
    0.9,
    1.8,
]
elongation3_capacity = [
    1.2,
    1.1,
    1.9,
]
elongation4_capacity = [
    1.1,
    1.3,
    1.9,
]
elongation5_capacity = [
    1.2,
    1.2,
    2.3,
]
average_temperature = [28.929, 14.026, 15.577,]

bar = (
    Bar(init_opts=opts.InitOpts(width="1680px", height="800px"))
    .add_xaxis(xaxis_data=x_data)
    .add_yaxis(
        series_name="第一次伸長量",
        yaxis_data=elongation1_capacity,
        yaxis_index=1,
        color=colors[0],
    )
    .add_yaxis(
        series_name="第二次伸長量",
        yaxis_data=elongation2_capacity,
        yaxis_index=1,
        color=colors[0],
    )
    .add_yaxis(
        series_name="第三次伸長量",
        yaxis_data=elongation3_capacity,
        yaxis_index=1,
        color=colors[0],
    )
    .add_yaxis(
        series_name="第四次伸長量",
        yaxis_data=elongation4_capacity,
        yaxis_index=1,
        color=colors[0],
    )
    .add_yaxis(
        series_name="第五次伸長量", 
        yaxis_data=elongation5_capacity, 
        yaxis_index=1, 
        color=colors[0]
    )
    .extend_axis(
        yaxis=opts.AxisOpts(
            name="伸長量",
            type_="value",
            min_=0,
            max_=6.0,
            position="right",
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(color=colors[0])
            ),
            axislabel_opts=opts.LabelOpts(formatter="{value} cm"),
        )
    )
    .extend_axis(
        yaxis=opts.AxisOpts(
            type_="value",
            name="力常數",
            min_=0,
            max_=30,
            position="left",
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(color=colors[1])
            ),
            axislabel_opts=opts.LabelOpts(formatter="{value} gw/cm"),
            splitline_opts=opts.SplitLineOpts(
                is_show=True, linestyle_opts=opts.LineStyleOpts(opacity=1)
            ),
        )
    )
    
 .set_global_opts(
        
        tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
    )
)


line = (
    Line()
    .add_xaxis(xaxis_data=x_data)
    .add_yaxis(
        series_name="力常數", y_axis=average_temperature, yaxis_index=2, color=colors[2]
    )
)
bar.overlap(line).render("自探.html")
#bar.overlap(line).render_notebook()
```
